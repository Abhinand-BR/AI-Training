```python
#probability distribution
#Discrete distribution describes outcomes that take on finite/countable values
#types 
# 1) Bernoulli's  distribution
# 2) binary distribution
# 3)poison distribution

#continuos distribution
#these distribution data that can take any value in the range
#types
#  1)uniform distribution
#  2)normal distribution
#  3)exponential distribution
```

```python
#statics - 
#central tendency - mean, median, mode,variance, standard deviation,range,interquartile range
#despersion - hypothesis testing,Anova

#bernoulli distributions - P^x(1-P)^(1-x)
from scipy. stats import bernoulli
import matplotlib.pyplot as plt 
P=0.7 #probability of success
x=[0,1]#probability outcomes
#probability mass function (pmf)
#probability density function(pmf)
Pmf = bernoulli.pmf(P,x)
plt.scatter(x,Pmf,color = "red")
plt.title("Bernoulli Distribution")
plt.xlabel("Bernoulli's Pmf value")
plt.ylabel("The possible outcomes")
plt.grid()
```

```python
#binomial distribution - nc^xP^xq^(n-x)
'''
n,p,q,x
''' 
from  scipy.stats import binom
import numpy as np
n = 10
p = 0.7 #q = 1-P
x = np.arange(0,n+1)
Pmf = binom.pmf(x,n,p)
print(Pmf)
plt.scatter(x,Pmf,color = "red")
plt.xlabel("The pmf value is ")
plt.ylabel("The outcome is")
plt.title("Binomial distribution") 
plt.show()
```

```python
#poission distribution
#Poission Distribution - e^-lambda.lambda^k/n!
'''
To calculate the average possible outcomes or rates with in the interval of time
e^-lambda.lambda^k/n!
''' 
from scipy.stats import poisson 
lambda_value = 5
k = np.arange(0,10)
Pmf = poisson.pmf(lambda_value,k)
plt.bar(k,Pmf,color = "black")
plt.show()
```

```python
#1 uniform didtribution - 1/b-a
from scipy.stats import uniform
a=2
b=5
x=np.linspace(a,b,200)
pdf = uniform.pdf(x,loc=a,scale=b-a)
print(pdf)
plt.bar(x,pdf,color = "orange")
plt.show()
```

```python
#exponential - average rate(lambda.e^-lambda)
from scipy.stats import expon
lambda_value = 2
x = np.linspace(0,10,100)
pdf = expon.pdf(x,scale = 1/lambda_value)
print(pdf)
plt.bar(x,pdf,color = "red")
plt.show()
```
